```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.worldometers.info/world-population/population-by-country/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html.parser')
```

```python
rows = soup.find('table',{'id':'example2'}).find('tbody').find_all('tr')
len(rows)
```

```python
countries_list = []
for i in ((rows)):
    dic = {}
    dic['Country'] = i.find_all('td')[1].text
    dic['Population (2023)'] = i.find_all('td')[2].text
    countries_list.append(dic)
countries_list
```

```python
df = pd.DataFrame(countries_list)
df.to_excel("countries_list.xlsx", index = True)
df.to_csv("countries_list.csv",index = False)
```
